```python
import random
import time
```

```python
min_temp = int(input("Enter minimum temperature limit: "))
max_temp = int(input("Enter maximum temperature limit: "))
```

```python
print("Starting Temperature Monitoring System...")
```

```python
while True:
    temp = random.randint(min_temp - 10, max_temp + 10)
    print(f"Current Temperature: {temp}°C")
    if temp < min_temp:
        print("Temperature too low!\n")
    elif temp > max_temp:
        print("Temperature too high!\n")
    else:
        print("Temperature is normal.\n")
    time.sleep(2)
```
